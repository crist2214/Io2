# Librería Sympy
SymPy es una librería de Python para matemáticas simbólicas. A continuación, exploraremos algunos de los conceptos básicos, como la declaración de símbolos, álgebra básica, cálculo y solución de ecuaciones.

## 1. Declaración de Símbolos

Para usar SymPy, primero debemos declarar las variables simbólicas con las que vamos a trabajar:

```python
from IPython.display import display, Markdown
from sympy import symbols # importo sólo lo que necesito (mejor que "import sympy as sym" en términos de consumo de memoria)

# Declarar variables simbólicas
x, y, z = symbols('x y z')

# Ejemplo
expresion = x + y + z
print("Expresión:", expresion)
```

## 2. Simplificación de Expresión

```python
from sympy import simplify

# Expresión a simplificar
expresion = (x**2 + 2*x + 1) / (x + 1)

# Simplificar la expresión
simplificada = simplify(expresion)
print("Expresión simplificada:", simplificada)
```

## 3. Derivadas

```python
from sympy import diff

# Derivar la función
f = x**3 + 2*x**2 + x
df = diff(f, x)

# simplemente escribimos la variable donde se guardó la derivada
df
```

## 4. Integrales

```python
from sympy import integrate

# Integral indefinida
g = x**2
int_g = integrate(g, x)
print(int_g)

# Integral definida entre 0 y 2
integral_definida = integrate(g, (x, 0, 2))
print("Integral definida:", integral_definida)
```

## 5. Solución de una ecuación

```python
from sympy import Eq, solve

# Definir una ecuación
ecuacion = Eq(x**2 + 2*x - 3, 0)

# Resolver la ecuación
soluciones = solve(ecuacion, x)
print("Soluciones:", soluciones)
```

## 6. Sistemas de Ecuaciones

```python
# Definir un sistema de ecuaciones
ecuacion1 = Eq(2*x + y, 1)
ecuacion2 = Eq(x - y, 3)

# Resolver el sistema
solucion_sistema = solve((ecuacion1, ecuacion2), (x, y))
print("Soluciones del sistema:", solucion_sistema)
```

## 7. Series de Taylor

```python
# Serie de Taylor de la función sin(x) en torno a x = 0
from sympy import sin

taylor = sin(x).series(x, 0, 6)
print("Serie de Taylor de sin(x):", taylor)
```

## 8. Gráficos con Sympy

```python
from sympy.plotting import plot

# Graficar la función sin(x)
plot(sin(x), (x, -10, 10), title="Gráfico de sin(x)")
```
